# Detecting Significant Impact Events

In this notebook we investigate methods to detect significant impact events. These are defined as events that exceed some threshold based on either linear acceleration or angular velocity.

```python
# Step 1: Import needed packages
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Read datalog CSV
df = pd.read_csv('../datalog.csv', index_col=False)
```

## Datalog Description

`../datalog.csv` contains a minute's worth of sensor data from the ICM20649 Accelerometer+Gyroscope sensor. It samples at a rate of 4500 Hz, and an antialiasing filter is applied on the sensor with a cutoff frequency of 1209 Hz.
